```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("brain_tumor_dataset.csv")
dataset.head(4)
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
sns.heatmap(dataset.isnull())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set figure size for better visibility
plt.figure(figsize=(12, 6))

# Create heatmap with a high-contrast colormap
sns.heatmap(dataset.isnull(), cmap="viridis", cbar=False)

# Add title for better understanding
plt.title("Missing Values Heatmap", fontsize=14)
 
# Show the plot
plt.show()
```
